```python
import pandas as pd
```

```python
expert_df = pd.read_csv("/home/luanagbmartins/Documents/CEIA/offline_to_online/CORL/data/results_expert.csv")
expert_df = expert_df[["env", "model", "episode_rewards_mean", "episode_rewards_std"]].groupby(["env", "model"]).mean().reset_index()
expert_df
```

```python
offline_df = pd.read_csv("/home/luanagbmartins/Documents/CEIA/offline_to_online/CORL/data/results_offline.csv")
offline_df["dataset_level"] = offline_df["dataset_id"].apply(lambda x: x.split("-")[-2])
offline_df = offline_df[["env", "dataset_level", "command_type", "model", "episode_rewards_mean", "episode_rewards_std"]].sort_values(["env", "dataset_level"])
offline_df
```

```python
random_df = pd.read_csv("/home/luanagbmartins/Documents/CEIA/offline_to_online/CORL/data/results_random.csv")
random_df = random_df[["env", "model", "episode_rewards_mean", "episode_rewards_std"]].groupby(["env", "model"]).mean().reset_index()
random_df["dataset_level"] = "random"
random_df
```

```python
combined_df = pd.concat([expert_df, random_df, offline_df], ignore_index=True)
combined_df = combined_df.sort_values(["env", "dataset_level"])
combined_df
```
